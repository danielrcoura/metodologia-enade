```python
## Filtrar cursos com conceito 5
conceito_5 = dados_conceito[dados_conceito['CONCEITO'] == '5']
# Filtrar alunos desses cursos
alunos_conceito_5 = dados[dados['CO_CURSO'].isin(conceito_5['CO_CURSO'])]
per_conceito5 = alunos_conceito_5.groupby(['QE_I08']).agg(['count']).apply(lambda x: x/1265*100, axis=1)
per_conceito5
```

```python
# Filtrar cursos com conceito 3
conceito_3 = dados_conceito[dados_conceito['CONCEITO'] == '3']
# Filtrar alunos desses cursos
alunos_conceito_3 = dados[dados['CO_CURSO'].isin(conceito_3['CO_CURSO'])]
per_conceito3 = alunos_conceito_3.groupby(['QE_I08']).agg(['count']).apply(lambda x: x/2465*100, axis=1)
per_conceito3
```

```python
labels = ['Até 1,5', '1,5 a 3', '3 a 4,5', '4,5 a 6', '6 a 10', '10 a 30', 'Acima de 30']

x = np.arange(len(labels))
width = 0.3

fig, ax = plt.subplots(1, 1, figsize=(12,12))
rects1 = ax.bar(x - width/2, [10.99, 24.67,23.49,13.79,15.29,10.50,1.26], width, color='limegreen', label='Curso com conceito 3')
rects2 = ax.bar(x + width/2, [7.43, 14.70,18.89,14.94,22.77,19.58,1.58], width, color='b', label='Curso com conceito 5')

ax.set_xlabel('Renda familiar (por salário mínimo)')
ax.set_ylabel('Percentual de alunos')
ax.set_title('Relação entre renda do aluno com o conceito do curso', size=20)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(fontsize = 16)
plt.xticks(size = 14)
plt.yticks(size = 14)


# conceito_3_bar = plt.bar(per_conceito3['NU_ANO'], 7)
# conceito_5_bar = plt.bar(7, per_conceito5['NU_ANO'], 0.2, color='b', label='Conceito 5')

# plt.xticks(3.6, ['a', 'b','c', 'd','e'])
# plt.legend()
```
